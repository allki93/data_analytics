```python
import numpy as np
import matplotlib.pyplot as plt
```

- простые линейные графики

```python
# 1-мерный массив (вектор-строка)
x_arr = np.arange(10)
x_arr
```

```python
y_arr = x_arr

# визуализация
plt.plot(x_arr, y_arr)

plt.show()
```

```python
# совмещение нескольких графиков в одном окне
y_arr_1 = x_arr ** 2
y_arr_2 = x_arr ** 3

# визуализация
plt.plot(x_arr, y_arr_1)
plt.plot(x_arr, y_arr_2)

plt.show()
```

```python
# 1-мерный массив из 100 значений
# линейной последовательности в диапозоне от -10 до 10
x_arr_1 = np.linspace(-10, 10, 100)
x_arr_1
```

```python
# функция синуса
y_1 = np.sin(x_arr_1)

# функция гиперболического тангенса
y_2 = np.tanh(x_arr_1)

# функция сигмоиды
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_3 = sigmoid(x_arr_1)

plt.plot(x_arr_1, y_1)
plt.plot(x_arr_1, y_2)
plt.plot(x_arr_1, y_3)

plt.show()
```

```python
# цвета линий

y_1 = np.sin(x_arr_1)
y_2 = np.sin(x_arr_1 + 1)
y_3 = np.sin(x_arr_1 + 2)
y_4 = np.sin(x_arr_1 + 3)

# по названию цвета
plt.plot(x_arr_1, y_1, color="red")
# по первой букве названия цвета из формата RGB или CMYK
plt.plot(x_arr_1, y_2, c='m')
# по 16-ричной кадировке RGB или RGBA
plt.plot(x_arr_1, y_3, c="#f97aa2")
# по формату RGB или RGBA в виде кортежа со значениями от 0 до 1
plt.plot(x_arr_1, y_4, c=(0.0, 0.8, 0.7))

plt.show()
```

```python
# стили линий

y_1 = np.sin(x_arr_1)
y_2 = np.sin(x_arr_1 + 1)
y_3 = np.sin(x_arr_1 + 2)
y_4 = np.sin(x_arr_1 + 3)

# сплошная линия
plt.plot(x_arr_1, y_1, linestyle="solid")
# штриховая линия
plt.plot(x_arr_1, y_2, c='m', ls="dashed")
# пунктирная линия
plt.plot(x_arr_1, y_3, c="#f97aa2", ls="dotted")
# штрих-пунктирная линия
plt.plot(x_arr_1, y_4, c=(0.0, 0.8, 0.7), ls="dashdot")

plt.show()
```

```python
# объединение настроек цвета и стиля линий

# штриховая зеленый
plt.plot(x_arr_1, y_1, "--g")
# пунктирная синяя
plt.plot(x_arr_1, y_2, ":b")
# штрих-пунктирная желтая
plt.plot(x_arr_1, y_3, "-.y")
# "звездная" красная
plt.plot(x_arr_1, y_4, "*r")

plt.show()
```

```python
# заголовок, легенда графика, подписи осей

# размер окна графика
plt.figure(figsize=(10, 7))

plt.plot(x_arr_1, np.sin(x_arr_1), "or", label="синус")
plt.plot(x_arr_1, np.cos(x_arr_1), "--g", label="косинус")

plt.legend(["синус", "косинус"])

plt.xlabel("x")
plt.ylabel("sin & cos")

plt.title("графики синуса и косинуса")

# диапазон координатной сетки
plt.axis([-15, 15, -2, 1])

plt.show()
```
